## Graph Problem

Evaluation Performance on Synthetic Graphs

```python
data = pd.read_csv("synthetic_graphs_nodes_400_small_gamma.csv")
data_random = data.loc[df['algorithm'] == 'GraphRandom']
data_greedy = data.loc[df['algorithm'] == 'GraphGreedy']
data_ranking = data.loc[df['algorithm'] == 'GraphRanking']
data_ground = data.loc[df['algorithm'] == 'GroundTruth']

score_random = data_random['score']
score_greedy = data_greedy['score']
score_ranking = data_ranking['score']
score_ground = data_ground['score']

skills = data_random['skills']
```

```python
plt.plot(skills, score_random, marker='o', color='g', label='GraphAlgoRand',linewidth=3)
plt.plot(skills, score_greedy, marker='o', color='orange', label='GraphAlgoGreedy',linewidth=3)
plt.plot(skills, score_ranking, marker='o', color='magenta', label='GraphAlgoRanking',linewidth=3)
plt.plot(skills, score_ground, marker='o', color='red', label='GroundTruth',linewidth=1)
plt.title('Synthetic Graph small gamma')
plt.ylabel('Score')
plt.xlabel('Skills')
plt.legend()
plt.show()
```
